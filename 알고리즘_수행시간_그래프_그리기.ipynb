```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,10))

# n 데이터 생성 (1부터 100까지) -> log그래프를 그려야 하는데 log그래프는 log_2(0)값은 없기 때문에 n값이 1일때부터로 지정함
x = np.linspace(1, 100, 1500)
# y축 범위 (0부터 3000까지)
plt.ylim(0, 3000)

# 2^n
y = [np.power(2, i) for i in x]
plt.plot(x, y, label='2^n')

# n^3
plt.plot(x, x**3, label='n^3')

# n^2
plt.plot(x, x**2, label='n^2')

# n*log_2^n
y = [i*np.log2(i) for i in x]
plt.plot(x, y, label='n*log_2(n)')

# n
plt.plot(x, x, label='n')

# log_2^n
y = [np.log2(i) for i in x]
plt.plot(x, y, label='log_2(n)')

plt.xlabel('n')
plt.ylabel('Time')

# 범례
legend_box = plt.legend(bbox_to_anchor=(1, 1))
frame = legend_box.get_frame()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20, 6))
# x축 데이터 생성 (0부터 100까지)
x = np.linspace(0, 20, 1400)

# n^3그래프
plt.plot(x, x**3, label='n^3')

# n^2그래프
plt.plot(x, x**2, label='n^2')

# n*log_2^n 그래프
y = [i*np.log2(i) if i > 0 else 1 for i in x] # log_2(0)은 정의되지 않으므로 예외 처리.
plt.plot(x, y, label='n*log_2(n)')

# n그래프 (연두색)
plt.plot(x, x, label='n', color='limegreen')

# log_2^n그래프
y = [np.log(i) if i > 0 else -np.inf for i in x] # log(0)은 -무한대이므로 예외 처리.
plt.plot(x, y, label='log(n)')

# 2^n 그래프
y = [np.power(2, i) for i in x]
plt.plot(x, y, label='2^n')

# y=1 그래프 추가
plt.axhline(y=1, color='red', linestyle='-', label='y=1')

# x축 및 y축 레이블 설정
plt.xlabel('n')
plt.ylabel('Time Complexity')

# 범례 추가 (밖으로 빼기)
legend_box = plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

frame = legend_box.get_frame()
frame.set_facecolor('white')
frame.set_edgecolor('black')

for text in legend_box.get_texts():
    text.set_color('black')

# y축 범위 설정 (0부터 1500까지)
plt.ylim(0, 10)

# 그림 크기 조절 (가로 길이 늘리기)


# 그림 출력
plt.tight_layout()
plt.show()
```
